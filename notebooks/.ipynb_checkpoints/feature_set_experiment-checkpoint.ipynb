```python
# autoreload chaque module spécifié par %aimport
%autoreload 1

%aimport src.runUtil

from collections import OrderedDict
from IPython.display import display
from os import path

import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping, ModelCheckpoint
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import CategoricalAccuracy
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import RMSprop

from src import constants
import src.runUtil as runUtil

 
```

```python
def load_feature_set(name):
    file_path = path.join(constants.RAW_TAGGED_FEATURE_SET_PATH, "msd-" + name + "/msd-" + name + ".csv")
    
    ftrs = np.array(pd.read_csv(file_path, header=None).values[:,2:-1])
    lbls = np.array(pd.read_csv(file_path, header=None).values[:,-1])

    return ftrs, lbls
    
def nn_proprocess(ftrs, lbls):
    lbls_1d = lbls.reshape(len(lbls), 1)
        
    oh_encoder = OneHotEncoder(sparse=False)
    lbls_oh = oh_encoder.fit_transform(lbls_1d)
    
    scaler = StandardScaler()
    ftrs_scld = scaler.fit_transform(ftrs)
    
    return ftrs_scld, lbls_oh

def run_experiment(model, run_name, x, y, epochs=1, batch_size=None):

    model_save_path = constants.MODELS_PATH + run_name + '.h5'
    
    tk_board = TensorBoard(log_dir=constants.LOGS_PATH + run_name)
    model_ckeckpt = ModelCheckpoint(filepath=model_save_path, monitor='val_accuracy', save_best_only=True, verbose=0)
    
    callbacks = [
        tk_board,
        model_ckeckpt
    ]
    
    history = model.fit(
        x=x, y=y, epochs=epochs, shuffle=True,
        batch_size=batch_size, validation_split=0.2,
        callbacks=callbacks, verbose=1
    )
    
    runUtil.save_run_history(run_name, history.history)

    hist_resume = runUtil.resume_history(history.history)
    
    run_results = OrderedDict([
        ('run_name', run_name)
    ])
    
    run_results.update(hist_resume)
    runUtil.save_run_results(run_name, run_results)
    
    config = {'epochs': epochs, 'batch_size': batch_size}
    runUtil.save_run_config(run_name, config)
    
    return run_results

def display_single_run_results(run_results):
    
    display(pd.DataFrame([run_results]).transpose())
```

## JMIR Derivatives
jmirderivatives

```python
ftr_jmirderivatives, lbl_jmirderivatives = load_feature_set('jmirderivatives_dev')
ftr_jmirderivatives, lbl_jmirderivatives_oh = nn_proprocess(ftr_jmirderivatives, lbl_jmirderivatives)

print(ftr_jmirderivatives.shape)
print(lbl_jmirderivatives_oh.shape)
```

```python
def create_jmirderivatives_mlp_model(input_shape):
    
    model = Sequential()
    
    model.add(Dense(90, activation='relu', input_shape=input_shape))
    model.add(Dense(60, activation='relu'))
    model.add(Dense(45, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(25, activation='softmax'))
    
    model.compile(
        optimizer=RMSprop(), 
        loss='categorical_crossentropy', 
        metrics=[
            CategoricalAccuracy(name='accuracy')
        ]
    )
    
    return model

jmirderivatives_input_shape = ftr_jmirderivatives[0].shape
jmirderivatives_mlp_model = create_jmirderivatives_mlp_model(jmirderivatives_input_shape)

jmirderivatives_mlp_results = run_experiment(
    jmirderivatives_mlp_model,
    'mlp_jmirderivatives_1',
    x=ftr_jmirderivatives,
    y=lbl_jmirderivatives_oh,
    epochs=500,
    batch_size=3000
)
```

```python
display_single_run_results(jmirderivatives_mlp_results)
```
